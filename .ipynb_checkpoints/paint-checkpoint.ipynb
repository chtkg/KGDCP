```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl  
from matplotlib.font_manager import FontProperties #字体管理器

font_times = FontProperties(fname="./font/times.ttf", size=15)

%matplotlib inline

mpl.rcParams['font.sans-serif']=['SimSun'] #用来正常显示中文标签  
# mpl.rcParams['font.sans-serif']=['Times New Roman'] #用来正常显示中文标签  
mpl.rcParams['axes.unicode_minus']=False #用来正常显示负号 
```

### 1、数据集各并发症人数分布图

```python
data = pd.read_excel('./data/complications_final.xlsx')
data.dropna(inplace=True)

diseases = ['脑梗', '冠心病', '视网膜病变', '糖尿病肾病', '周围神经病变']
sick_humans = [sum(data[disease]) for disease in diseases]
not_sick_humans = [data[disease].size - sum(data[disease]) for disease in diseases]

X1 = [i - 0.2 for i in range(len(diseases))]
X2 = [i + 0.2 for i in range(len(diseases))]

plt.figure(figsize=(8, 5))
plt.rc('font', size=12) # controls default text sizes
# plt.rc('xtick', labelsize=11) # fontsize of the tick labels
plt.rc('ytick', labelsize=13) # fontsize of the tick labels
plt.rc('figure', titlesize=14) # fontsize of the figure title
plt.bar(X1,  sick_humans, width=0.4, label='患病', color='#ae8ec2')
plt.bar(X2,  not_sick_humans, width=0.4, label='未患病', color='#82b2cd')
plt.xticks([i for i in range(len(diseases))], diseases)
for x, y in zip(X1, sick_humans):
    plt.text(x, y, y, ha='center', va='bottom', fontproperties=font_times)
for x, y in zip(X2, not_sick_humans):
    plt.text(x, y, y, ha='center', va='bottom', fontproperties=font_times)
plt.ylabel('人数')
plt.legend()
plt.savefig('./image/数据集分布图.png', dpi=300)
plt.show()
```

### 2、整合前后知识表示学习模型MR值和Hit@10值

```python
mr_before = [811.2, 811.9, 1224.5]
hit10_before = [21.4, 24.6, 9.2]
mr_after = [776.3, 703.6, 1131.3]
hit10_after = [22.1, 26.9, 10.4]

labels = ['TransE', 'TransH', 'TransR'] 

X1 = [i for i in range(0, 3)]
X2 = [i+0.3 for i in range(0, 3)]

fig = plt.figure(figsize=(12, 4))

plt.rc('font', size=12) # controls default text sizes
plt.rc('xtick', labelsize=13) # fontsize of the tick labels
plt.rc('figure', titlesize=14) # fontsize of the figure title

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.bar(X1, mr_before, width=0.3, label='整合前', color='#58539f')
ax1.bar(X2, mr_after, width=0.3, label='整合后', color='#d86967')
ax1.set_title('MR', fontproperties='Times New Roman')
ax1.set_xticks([i+0.15 for i in range(0, 3)], labels, fontproperties='Times New Roman')
for x, y in zip(X1, mr_before):
    ax1.text(x-0.13, y+20, y, fontproperties='Times New Roman')
for x, y in zip(X2, mr_after):
    ax1.text(x-0.13, y+20, y, fontproperties='Times New Roman')
ax1.legend()

ax2.bar(X1, hit10_before, width=0.3, label='整合前', color='#58539f')
ax2.bar(X2, hit10_after, width=0.3,  label='整合后', color='#d86967')
ax2.set_title('Hit@10', fontproperties='Times New Roman')
ax2.set_xticks([i+0.13 for i in range(0, 3)], labels, fontproperties='Times New Roman')
for x, y in zip(X1, hit10_before):
    ax2.text(x, y, y, ha='center', va='bottom', fontproperties='Times New Roman')
for x, y in zip(X2, hit10_after):
    ax2.text(x, y, y, ha='center', va='bottom', fontproperties='Times New Roman')
ax2.legend()
plt.savefig('./image/知识整合前后对比图.png', dpi=300)
plt.show()
```

### 3、深度学习模型比较

```python
acc_ng = [0.6268, 0.6567, 0.6716, 0.6865, 0.7462]
recall_ng = [0.6388, 0.6580, 0.6402, 0.6313, 0.7152]
acc_zw = [0.5909, 0.5606,  0.6363, 0.5909, 0.6515]
recall_zw = [0.5666, 0.5166, 0.6194, 0.5527, 0.6444]

X1 = [i for i in range(0, 5)]
X2 = [i+0.3 for i in range(0, 5)]

labels = ['DNN', 'CNN', 'RNN', 'LSTM', 'KGDCP']

fig = plt.figure(figsize=(13, 7))
plt.rc('font', size=12) # controls default text sizes
# plt.rc('xtick', labelsize=11) # fontsize of the tick labels
plt.rc('ytick', labelsize=13) # fontsize of the tick labels
plt.rc('figure', titlesize=14) # fontsize of the figure title
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.barh(X1, acc_ng,  height=0.3, label='Accuracy', color='#9b95c9')
ax1.barh(X2, recall_ng,  height=0.3, label='Recall', color='#65c294')
ax1.set_title('脑梗')
ax1.set_yticks([i+0.15 for i in range(0, 5)], labels, fontproperties='Times New Roman')
for x, y in zip(X1, acc_ng):
    ax1.text(0.02, x-0.05, '{:.2f}%'.format(y*100), fontproperties='Times New Roman')
for x, y in zip(X2, recall_ng):
    ax1.text(0.02, x-0.05, '{:.2f}%'.format(y*100), fontproperties='Times New Roman')
ax1.legend()

ax2.barh(X1, acc_zw, height=0.3, label='Accuracy', color='#9b95c9')
ax2.barh(X2, recall_zw, height=0.3, label='Recall', color='#65c294')
ax2.set_title('周围神经病变')
ax2.set_yticks([i+0.15 for i in range(0, 5)], labels, fontproperties='Times New Roman')
for x, y in zip(X1, acc_zw):
    ax2.text(0.02, x-0.05, '{:.2f}%'.format(y*100), fontproperties='Times New Roman')
for x, y in zip(X2, recall_zw):
    ax2.text(0.02, x-0.05, '{:.2f}%'.format(y*100), fontproperties='Times New Roman')
ax2.legend()
plt.savefig('./image/深度学习模型比较.png', dpi=300)
plt.show()

```
